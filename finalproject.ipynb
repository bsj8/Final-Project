**The best neighborhood is the neighborhood with the most speed bumps and road markings. This keeps drivers safe creating a happier, healthier environment. The speed bumps and markings mae accidents happend less often making the best neighborhood a great place to live.**

```python
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

import pandas as pd
import numpy as np
%matplotlib inline

neighborhoods_with_bumps = ["Squirrel Hill South", "Polish Hill", "North Oakland", "Upper Hill", "Homewood South", "Brookline", "Squirrel Hill North", "South Side Flats", "Mount Washington", "Bloomfield", "Highland Park", "Carrick", "Central Oakland"]
```

```python
speed_humps = pd.read_csv("https://data.wprdc.org/datastore/dump/37b2ac41-ae8e-4de1-8405-157e05dc3640", index_col = "id")
df = speed_humps[["neighborhood","speed_hump_material"]]
speed_humps.head(100)
```

```python
df = df.sort_values(by="neighborhood")
df.head(100)
```

```python
df = df['neighborhood'].value_counts().sort_values(ascending=False)
df
```
